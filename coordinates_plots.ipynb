```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# riders
df = pd.read_excel("riders.xlsx")

df1 = df.copy()
df1[['x', 'y']] = df1['pickup_location'].str.strip('()').str.split(',', expand=True).astype(float)
df2 = df.copy()
df2[['x', 'y']] = df2['dropoff_location'].str.strip('()').str.split(',', expand=True).astype(float)

# drivers
df = pd.read_excel("drivers.xlsx")
df3 = df.copy()
df3[['x', 'y']] = df3['initial_location'].str.strip('()').str.split(',', expand=True).astype(float)
df4 = df.copy()
df4[['x', 'y']] = df4['current_location'].str.strip('()').str.split(',', expand=True).astype(float)
```

```python
# Scatter plot next to each other
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(
    df1['x'],
    df1['y'],
    s=2,
    alpha=0.3 
)
plt.xlabel('Longitude / X')
plt.ylabel('Latitude / Y')
plt.title('Riders : Pickup Locations Scatter Plot')

plt.subplot(1, 2, 2)
plt.scatter(
    df2['x'],
    df2['y'],
    s=2,
    alpha=0.3 
)
plt.xlabel('Longitude / X')
plt.ylabel('Latitude / Y')
plt.title('Riders : Dropoff Locations Scatter Plot')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Lines
for i in range(len(df1)):
    ax[0].plot(
        [df1.loc[i, 'x'], df2.loc[i, 'x']],
        [df1.loc[i, 'y'], df2.loc[i, 'y']],
        color='blue',
        alpha=0.1
    )

ax[0].set_title('Pickup to Dropoff')
ax[0].set_xlabel('Longitude / X')
ax[0].set_ylabel('Latitude / Y')

# Midpoints
mid_x = (df1['x'] + df2['x']) / 2
mid_y = (df1['y'] + df2['y']) / 2

ax[1].scatter(mid_x, mid_y, s=2, alpha=0.3)

ax[1].scatter(
    mid_x.mean(),
    mid_y.mean(),
    color='red',
    s=30,
    label='Centroid'
)

plt.text(
    mid_x.mean(),
    mid_y.mean(),
    f'({mid_x.mean():.4f}, {mid_y.mean():.4f})',
    fontsize=10,
    ha='center',
    va='bottom'
)
# print(mid_x.mean(), mid_y.mean())

ax[1].legend()
ax[1].set_title('Midpoints')
ax[1].set_xlabel('Longitude / X')
ax[1].set_ylabel('Latitude / Y')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot
plt.figure()
plt.scatter(
    df3['x'],
    df3['y'],
    s=2,
    alpha=0.3 
)

plt.xlabel('Longitude / X')
plt.ylabel('Latitude / Y')
plt.title('Driver : Inital Locations Scatter Plot')

plt.show()
```

```python
# plot points by distinguish in two different colors of
# 1) if drivers current location matches with riders dropoff location
# 2) if drivers current location does not match with riders dropoff location

count = 0
for i in range(len(df4)):
    if any((df4.loc[i, 'x'] == df2['x']) & (df4.loc[i, 'y'] == df2['y'])):
        plt.scatter(df4.loc[i, 'x'], df4.loc[i, 'y'], color='green', s=5, alpha=0.5)
    else:
        plt.scatter(df4.loc[i, 'x'], df4.loc[i, 'y'], color='red', s=10, alpha=1.0)
        count += 1
plt.xlabel('Longitude / X')
plt.ylabel('Latitude / Y')
plt.title('Driver Locations: Green = Matches Dropoff, Red = No Match')
plt.show()

print(f"Number of drivers matching riders dropoff locations: {len(df4) - count}")
print(f"Number of drivers NOT matching riders dropoff locations: {count}")
```
