```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("riders.xlsx")

df[['x', 'y']] = df['pickup_location'].str.strip('()').str.split(',', expand=True).astype(float)

# Scatter plot
plt.figure()
plt.scatter(
    df['x'],
    df['y'],
    s=2,
    alpha=0.3 
)

plt.xlabel('Longitude / X')
plt.ylabel('Latitude / Y')
plt.title('Riders : Pickup Locations Scatter Plot')

plt.show()
```
